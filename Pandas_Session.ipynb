# Logical Operators in Pandas:
    * & (and)
    * | (or)
    * ~ (not)    tilde

```python
# In Pandas show the sign irrespective of writing all as we did in python.
```

```python
# SMID = Single Instruction and Multiple Data Explore.
```

```python
# As your given instructions will just be single, so no if else statement is used in Pandas.
```

```python
df['State'] == 'Florida'
```

```python
df[df['State'] == 'Florida']
```

```python
df['Category'].unique()
```

```python
df[df['Category'] == 'Technology']
```

```python
df[(df['State'] == 'Florida') & (df['Category'] == 'Technology')]
```

```python
df[(df['State'] == 'Florida') | (df['State'] == 'California')]
```

```python
(df['State'] == 'Florida') ~(df['Category'] == 'Technology')
```

```python
(df['State'] == 'Florida') & ~(df['Category'] == 'Technology')
```

```python
df[(df['State'] == 'Florida') & ~(df['Category'] == 'Technology')]
```

```python
df[(df['State'] == 'Florida') | ~(df['Category'] == 'Technology')]
```
